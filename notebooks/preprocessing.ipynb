```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd

demo = pd.read_csv('../data/train_OwBvO8W/demographics.csv')
print(demo.shape)
demo.head()
```

```python
event = pd.read_csv('../data/train_OwBvO8W/event_calendar.csv')
print(event.shape)
event.head()
```

```python
historical = pd.read_csv('../data/train_OwBvO8W/historical_volume.csv')
print(historical.shape)
historical.head()
```

```python
soda = pd.read_csv('../data/train_OwBvO8W/industry_soda_sales.csv')
print(soda.shape)
soda.head()
```

```python
industry = pd.read_csv('../data/train_OwBvO8W/industry_volume.csv')
print(industry.shape)
industry.head()
```

```python
price = pd.read_csv('../data/train_OwBvO8W/price_sales_promotion.csv')
print(price.shape)
price.head()
```

```python
weather = pd.read_csv('../data/train_OwBvO8W/weather.csv')
print(weather.shape)
weather.head()
```

```python
sku = historical.merge(price,on=['Agency', 'SKU', 'YearMonth'],how='left')
sku = sku.merge(soda, on=['YearMonth'], how='left')
sku = sku.merge(industry, on='YearMonth', how='left')
sku = sku.merge(weather, on=['YearMonth', 'Agency'], how='left')
sku = sku.merge(event,on=['YearMonth'], how='left')
print(sku.shape)
sku.head()
```

```python
sku = sku.drop(['FIFA U-17 World Cup', 'Football Gold Cup'], axis=1)
sku['Month'] = sku['YearMonth'].apply(lambda x: str(x)[-2:])
sku.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, PowerTransformer, StandardScaler

month_onehot = pd.DataFrame(OneHotEncoder().fit_transform(sku[['Month']]).toarray(), columns=list(map(lambda x: 'Month%.2d' % x, range(1, 13))))
sku = pd.concat([sku, month_onehot], axis=1).drop(['Month'], axis=1)
sku.head()
```

```python
pwr_cols = ['Volume', 'Price', 'Sales', 'Promotions', 'Avg_Max_Temp']
std_cols = ['Soda_Volume', 'Industry_Volume']
pwr_scaler = PowerTransformer().fit(sku[pwr_cols])
std_scaler = StandardScaler().fit(sku[std_cols])

pwr_data = pd.DataFrame(pwr_scaler.transform(sku[pwr_cols]), columns=list(map(lambda x: f'tr_{x}', pwr_cols)))
std_data = pd.DataFrame(std_scaler.transform(sku[std_cols]), columns=list(map(lambda x: f'tr_{x}', std_cols)))
sku = pd.concat([sku, pwr_data, std_data], axis=1)
sku.head()
```

```python
num_cols = list(filter(lambda x: x.startswith('tr_'), sku.columns))
onehot_cols = sku.columns.to_list()[10:-7]
base_cols = ['Agency', 'SKU', 'YearMonth']

agencies = sorted(sku['Agency'].drop_duplicates())
skues = sorted(sku['SKU'].drop_duplicates())
months = sorted(sku['YearMonth'].drop_duplicates())
len(agencies), len(skues), len(months)
```

```python
def make_seq_data(data, target, sku, l_seq, given_list=None):
    target_data = data.loc[data['SKU'] == sku][['YearMonth', 'Agency', target]].sort_values(['Agency', 'YearMonth'])
    seq_data = target_data.pivot(index='Agency', columns='YearMonth', values=target)
    if seq_data.shape[0] != len(agencies): 
        dummy = pd.DataFrame(np.zeros((len(agencies) - seq_data.shape[0], seq_data.shape[1])), index=sorted(set(agencies) - set(seq_data.index)), columns=seq_data.columns)
        seq_data = pd.concat([seq_data, dummy], axis=0).sort_index()
    if seq_data.shape[1] != len(months):
        dummy = pd.DataFrame(np.zeros((seq_data.shape[0], len(months) - seq_data.shape[1])), index=seq_data.index, columns=sorted(set(months) - set(seq_data.columns)))
        seq_data = pd.concat([seq_data, dummy], axis=1)[agencies]
    if given_list is None:
        json_data = []
    else:
        json_data = given_list
    for i in range(l_seq, len(months) - 1):
        single = list(filter(lambda x: x['date'] == months[i], json_data))
        if len(single) == 0:
            single = {'date': months[i]}
        else:
            single = single[0]
        single[f'input_{target[4:]}'] = seq_data[months[i - l_seq:i]].values.tolist()
        single[f'date_total_{target[4:]}'] = seq_data[months[i - l_seq:i]].sum(0).values.tolist()

        if target == 'tr_Volume':
            single['output'] = seq_data[months[i + 1]].values.tolist()
        json_data.append(single)
    return json_data

def make_seq_data_by_date(data, target_name, target_list, l_seq, given_list=None):
    seq_data = data[['YearMonth'] + target_list].drop_duplicates().sort_values(['YearMonth'])
    seq_data.index = seq_data['YearMonth']
    seq_data = seq_data.drop(['YearMonth'], axis=1).T
    if given_list is None:
        json_data = []
    else:
        json_data = given_list
    for i in range(l_seq, len(months) - 1):
        single = list(filter(lambda x: x['date'] == months[i], json_data))
        if len(single) == 0:
            single = {'date': months[i]}
        else:
            single = single[0]
        single[target_name] = seq_data[months[i - l_seq:i]].values.tolist()
        json_data.append(single)
    return json_data
```

```python
import json 

l_seq = 5
for i, s in enumerate(skues):
    result_list = None
    for c in num_cols:
        result_list = make_seq_data(sku, c, s, l_seq, result_list)
    result_list = make_seq_data_by_date(sku, 'input_month', onehot_cols[-12:], l_seq, result_list)
    result_list = make_seq_data_by_date(sku, 'input_event', onehot_cols[:-12], l_seq, result_list)
    if i == 0:
        all_df = pd.DataFrame(result_list)
        test_idx = list(all_df[all_df['date'] >= 201701].index)
        train_val_idx = sorted(set(np.arange(len(all_df))) - set(test_idx))
        val_idx = np.random.choice(train_val_idx, round(len(train_val_idx) * 0.1), replace=False)
        train_idx = sorted(set(train_val_idx) - set(val_idx))
        print('!!', len(all_df), len(train_idx), len(val_idx), len(test_idx))
    train_list, val_list, test_list = [], [], []
    for j in range(len(result_list)):
        if j in train_idx:
            train_list.append(result_list[j]) 
        elif j in val_idx:
            val_list.append(result_list[j]) 
        else:
            test_list.append(result_list[j]) 
    print(len(result_list), len(train_list), len(val_list), len(test_list))
    with open(f'../sku_seq_data/train_{s}.json', 'w') as f:
        json.dump(train_list, f)
    with open(f'../sku_seq_data/val_{s}.json', 'w') as f:
        json.dump(val_list, f)
    with open(f'../sku_seq_data/test_{s}.json', 'w') as f:
        json.dump(test_list, f)
```
